```python
# fourth  moment  Business decision
```

```python
kurtosis_values = df.kurtosis()
kurtosis_values
```


Crop_Year: -1.135214

The negative kurtosis suggests that the distribution of "Crop_Year" is platykurtic, meaning it has lighter tails than a normal distribution. In simpler terms, the distribution has fewer extreme values (outliers) than a normal distribution.

Area: 5568.375426

The high positive kurtosis for "Area" (5568.375426) indicates that the distribution has heavier tails than a normal distribution. This suggests the presence of outliers or extreme values in the "Area" column.

Production: 1882.113472

Similarly, the high positive kurtosis for "Production" (1882.113472) indicates that the distribution has heavier tails than a normal distribution. This suggests the presence of outliers or extreme values in the "Production" column.

# Insights of Exploratory data analysis


The "Production" column contains null values, and both "Area" and "Production" exhibit outliers. Additionally, both "Production" and "Area" are positively skewed, indicating that the majority of data points fall within lower values, with fewer instances of higher production and area values. Importantly, both production and area used for cultivation are low. Predominantly, crops are cultivated during the Kharif season, and the primary crop grown is rice. Uttar Pradesh emerges as the leading producer, while Punjab holds the highest agricultural area, emphasizing the dominance of agricultural activities in the region.
